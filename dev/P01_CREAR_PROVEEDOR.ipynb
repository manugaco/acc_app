```python
# globally declare wb and sheet variable 
  
# opening the existing excel file
file='../P01_BBDD/DB_PROVEEDORES.xlsx'
wb = load_workbook(file) 
tipo='PROVEEDOR'

# create the sheet object 
sheet = wb.active 

# Function to take data from GUI  
# window and write to an excel file 
def insert(): 
      
    # if user not fill any entry 
    # then print "empty input" 
    if (
        nombre_field.get() == "" and
        razon_field.get() == "" and
        cif_field.get() == "" and
        cp_field.get() == ""
       ): 
              
        print("Rellenar Campos: NOMBRE, RAZON, CIF, CP.") 
  
    else: 
  
        # assigning the max row and max column 
        # value upto which data is written 
        # in an excel sheet to the variable 
        current_row = sheet.max_row 
        current_column = sheet.max_column 
  
        # get method returns current text 
        # as string which we write into 
        # excel spreadsheet at particular location 
        sheet.cell(row=current_row + 1, column=1).value = new_id(file, tipo)
        sheet.cell(row=current_row + 1, column=2).value = nombre_field.get() 
        sheet.cell(row=current_row + 1, column=3).value = razon_field.get() 
        sheet.cell(row=current_row + 1, column=4).value = cif_field.get() 
        sheet.cell(row=current_row + 1, column=5).value = direccion_field.get()
        sheet.cell(row=current_row + 1, column=6).value = cp_field.get()
        sheet.cell(row=current_row + 1, column=7).value = loc_field.get() 
        sheet.cell(row=current_row + 1, column=8).value = prov_field.get() 
        sheet.cell(row=current_row + 1, column=9).value = ccaa_field.get()
        sheet.cell(row=current_row + 1, column=10).value = pais_field.get() 
        sheet.cell(row=current_row + 1, column=11).value = tel_field.get() 
        sheet.cell(row=current_row + 1, column=12).value = mov_field.get() 
        sheet.cell(row=current_row + 1, column=13).value = fax_field.get() 
        sheet.cell(row=current_row + 1, column=14).value = email_field.get() 
        sheet.cell(row=current_row + 1, column=15).value = web_field.get()
  
        # save the file 
        wb.save(file) 
  
        # set focus on the name_field box 
        name_field.focus_set() 
  
        # call the clear() function 
        clear() 
  
  
# Driver code 
if __name__ == "__main__": 
      
    # create a GUI window 
    root = Tk() 
  
    # set the title of GUI window 
    root.title("Registro Proveedores") 
  
    # set the configuration of GUI window 
    root.geometry("1200x500") 
  
    excel() 
  
    # Etiquetas  
    ids = Label(root, text="ID" + ":")
    ids_t = Label(root, text=str(new_id(file, tipo)))
    nombre = Label(root, text="NOMBRE:")
    razon = Label(root, text="RAZON:")
    cif = Label(root, text="CIF:")
    direccion = Label(root, text="DIRECCION:")  
    cod_pos= Label(root, text="CODIGO POSTAL:")  
    local = Label(root, text="LOCALIDAD:")  
    prov = Label(root, text="PROVINCIA:",) 
    ccaa = Label(root, text="CCAA:")
    pais = Label(root, text="PAIS:") 
    tel = Label(root, text="TELEFONO:",)
    mov = Label(root, text="MOVIL:") 
    fax = Label(root, text="FAX:") 
    email = Label(root, text="EMAIL:") 
    webs = Label(root, text="WEB:")
  
    # create a text entry box 
    # for typing the information
    nombre_field = Entry(root) 
    razon_field = Entry(root)
    cif_field = Entry(root)
    direccion_field = Entry(root) 
    cp_field = Entry(root) 
    loc_field = Entry(root) 
    prov_field = Entry(root) 
    ccaa_field = Entry(root)
    pais_field = Entry(root) 
    tel_field = Entry(root) 
    mov_field = Entry(root) 
    fax_field = Entry(root) 
    email_field = Entry(root) 
    web_field = Entry(root) 
   
    # whenever the enter key is pressed 
    # then call the focus function
    nombre_field.bind("<Return>", focus2)  
    razon_field.bind("<Return>", focus3)
    cif_field.bind("<Return>", focus4)
    direccion_field.bind("<Return>", focus5) 
    cp_field.bind("<Return>", focus6)  
    loc_field.bind("<Return>", focus7)  
    prov_field.bind("<Return>", focus8)
    ccaa_field.bind("<Return>", focus9)  
    pais_field.bind("<Return>", focus10) 
    tel_field.bind("<Return>", focus11) 
    mov_field.bind("<Return>", focus12)  
    fax_field.bind("<Return>", focus13)  
    email_field.bind("<Return>", focus14)
    web_field.bind("<Return>", focus15)
  
    # grid method is used for placing 
    # the widgets at respective positions 
    # in table like structure .
    ids.grid(row=1, column=0, sticky="W") 
    ids_t.grid(row=1, column=1, sticky="W", pady=30)
    
    nombre.grid(row=2, column=0, sticky="W")
    nombre_field.grid(row=2, column=1, sticky="W", ipadx="100")
    
    razon.grid(row=2, column=2, sticky="W")
    razon_field.grid(row=2, column=3, sticky="W", ipadx="10") 
     
    cif.grid(row=2, column=4, sticky="W",pady=20)
    cif_field.grid(row=2, column=5, sticky="W", ipadx="10",pady=20)
    
    direccion.grid(row=5, column=0, sticky="W")
    direccion_field.grid(row=5, sticky="W", column=1, ipadx="100") 
    
    cod_pos.grid(row=5, column=2, sticky="W")
    cp_field.grid(row=5, column=3, sticky="W", ipadx="10")
    
    local.grid(row=7, column=0, sticky="W")
    loc_field.grid(row=7, column=1, sticky="W", ipadx="100")
    
    prov.grid(row=8, column=0, sticky="W")
    prov_field.grid(row=8, column=1, sticky="W", ipadx="100")
    
    ccaa.grid(row=9, column=0, sticky="W")
    ccaa_field.grid(row=9, column=1, sticky="W", ipadx="100")
    
    pais.grid(row=10, column=0, sticky="W")
    pais_field.grid(row=10, column=1, sticky="W", ipadx="100") 
    
    tel.grid(row=11, column=0, sticky="W")
    tel_field.grid(row=11, column=1, sticky="W", ipadx="50") 
    
    mov.grid(row=11, column=2, sticky="W")
    mov_field.grid(row=11, column=3, sticky="W", ipadx="50")
    
    fax.grid(row=11, column=4, sticky="W",pady=30)
    fax_field.grid(row=11, column=5, sticky="W", ipadx="50",pady=30)
    
    email.grid(row=12, column=0, sticky="W")
    email_field.grid(row=12, column=1, sticky="W", ipadx="100") 
    
    webs.grid(row=12, column=2, sticky="W",pady=30) 
    web_field.grid(row=12, column=3, sticky="W", ipadx="100",pady=30)
    
    # call excel function 
    excel() 
    
    # create a Submit Button and place into the root window 
    submit = Button(root, text="Crear Proveedor", fg="Black", command = insert)
    
    close = Button(root, text="Cerrar", fg="Black", command = close_window)
    
    submit.grid(row=15, column=2, pady=30)
    close.grid(row=15, column=3, pady=30)
  
    # start the GUI 
    root.mainloop()
```
